# RocksDB WAL Experiments Graph Generation
This notebook plots the data collected from the 5 RocksDB WAL experiments and saves the visual graphs into the `graphs/` folder.

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create graphs directory to save images
os.makedirs('graphs', exist_ok=True)
plt.style.use('ggplot')
print("Libraries loaded and graphs directory created.")
```

### Experiment 1: WAL Throughput Cost

```python
labels1 = ['WAL=ON\nbatch=1', 'WAL=OFF\nbatch=1', 'WAL=ON\nsync=ON', 'WAL=ON\nbatch=100', 'WAL=ON\nbatch=1000']
ops1 = [89590, 300333, 228, 281076, 313667]
plt.figure(figsize=(10, 6))
bars = plt.bar(labels1, ops1, color=['#3498db', '#e74c3c', '#2ecc71', '#9b59b6', '#f1c40f'])
plt.title('Experiment 1: WAL Throughput Cost')
plt.ylabel('Throughput (ops/s)')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5000, f'{yval:,}', ha='center', va='bottom')
plt.tight_layout()
plt.savefig('graphs/exp1_throughput.png')
plt.show()
```

### Experiment 2: Crash Recovery

```python
labels2 = ['Normal\n(SST Intact)', 'WAL=ON\n(SST Lost)', 'WAL=OFF\n(SST Lost)']
recov2 = [5000, 5000, 0]
plt.figure(figsize=(8, 5))
bars = plt.bar(labels2, recov2, color=['#2ecc71', '#3498db', '#e74c3c'])
plt.title('Experiment 2: Keys Recovered After Crash Sim')
plt.ylabel('Keys Recovered (Out of 5000)')
plt.ylim(0, 6000)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, f'{yval:,}', ha='center', va='bottom')
plt.tight_layout()
plt.savefig('graphs/exp2_recovery.png')
plt.show()
```

### Experiment 3: Behavior Under Skew

```python
# Exp 3A: Value Size Impact
labels3A = ['64B', '512B', '4KB', '64KB']
wal3A = [4, 25, 196, 246]
data3A = [3, 24, 195, 3125]
x3A = np.arange(len(labels3A))
width = 0.35
fig, ax = plt.subplots(figsize=(8, 5))
rects1 = ax.bar(x3A - width/2, wal3A, width, label='WAL Size (MB)', color='#e74c3c')
rects2 = ax.bar(x3A + width/2, data3A, width, label='Data Written (MB)', color='#34495e')
ax.set_ylabel('Megabytes (MB)')
ax.set_title('Experiment 3A: Value Size Impact on WAL (50k ops)')
ax.set_xticks(x3A)
ax.set_xticklabels(labels3A)
ax.legend()
plt.tight_layout()
plt.savefig('graphs/exp3_skew_A.png')
plt.show()

# Exp 3B: Hot Key
labels3B = ['Uniform Keys', 'Hot Key (1 key)']
thr3B = [72534, 96784]
fig2, ax2 = plt.subplots(figsize=(6, 5))
bars3B = ax2.bar(labels3B, thr3B, color=['#9b59b6', '#f1c40f'])
ax2.set_title('Experiment 3B: Throughput (Uniform vs Hot Key)')
ax2.set_ylabel('Throughput (ops/s)')
ax2.set_ylim(0, 110000)
for bar in bars3B:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval + 2000, f'{yval:,}', ha='center', va='bottom')
plt.tight_layout()
plt.savefig('graphs/exp3_skew_B.png')
plt.show()
```

### Experiment 4: Write Group Batching

```python
batch_sizes = [1, 10, 50, 100, 500, 1000, 5000, 10000]
thr4 = [82552, 214418, 267562, 258377, 269464, 264133, 257030, 258557]
labels4 = [str(b) for b in batch_sizes]
plt.figure(figsize=(10, 5))
plt.plot(labels4, thr4, marker='o', linestyle='-', color='#e67e22', linewidth=2, markersize=8)
plt.title('Experiment 4: Throughput vs Write Batch Size')
plt.xlabel('Batch Size')
plt.ylabel('Throughput (ops/s)')
plt.grid(True)
plt.tight_layout()
plt.savefig('graphs/exp4_batch_throughput.png')
plt.show()
```

### Experiment 5: Data Growth

```python
# Exp 5A: Ops Scale
ops5 = [1000, 5000, 10000, 50000, 100000]
rec5A = [48, 71, 89, 304, 561]
labels5A = [f'{op//1000}k' for op in ops5]
plt.figure(figsize=(8, 5))
plt.plot(labels5A, rec5A, marker='s', linestyle='-', color='#c0392b', linewidth=2, markersize=8)
plt.title('Experiment 5A: Recovery Time vs Ops Count (256B vals)')
plt.xlabel('Operations')
plt.ylabel('Recovery Time (ms)')
plt.grid(True)
plt.tight_layout()
plt.savefig('graphs/exp5_growth_ops.png')
plt.show()

# Exp 5B: Value Scale
vsize5 = ['64B', '512B', '4096B', '16384B']
rec5B = [77, 110, 370, 1114]
plt.figure(figsize=(8, 5))
plt.plot(vsize5, rec5B, marker='^', linestyle='-', color='#2980b9', linewidth=2, markersize=8)
plt.title('Experiment 5B: Recovery Time vs Value Size (10k ops)')
plt.xlabel('Value Size')
plt.ylabel('Recovery Time (ms)')
plt.grid(True)
plt.tight_layout()
plt.savefig('graphs/exp5_growth_vsize.png')
plt.show()
```
